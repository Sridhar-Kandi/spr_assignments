```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 5, 6]
y = [20, 40, 10, 6]
plt.plot(x, y)
plt.show()
```

```python
z = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri']
a = [200, 100, 1200, 170, 165]
plt.plot(z, a, color='red', linestyle='-', linewidth = 2, marker = 'o', label='Sales data 2025')
plt.xlabel('Days of the week')
plt.ylabel('Sales per week')
plt.title('Monthly sales data report')
plt.legend(fontsize=8)
plt.grid(color='grey', linestyle=':', linewidth = 1)
#plt.xlim(1, 4)
#plt.ylim(0, 2000)
plt.show()
```
